```python
#@title # 年度を計算する関数
def business_year(date):
    if date.month >= 4:
        return date.year
    else:
        return date.year - 1

date = pd.to_datetime('2025/3/6')
ret = business_year(date)
print(ret)
```

```python
#@title ## 期を算出（日付型の引数）
import datetime

# 期を算出
def ki(date):
    if date.year >= 2025 and date.month >=4:
        ki = date.year - 1978
    elif date.month < 10 :
        ki = date.year - 1979
    else :
        ki = date.year - 1978

    return ki

# 文字列を日付型に変換
tstr = '2025-12-29'
date = datetime.datetime.strptime(tstr, '%Y-%m-%d')
ki = ki(date)
print(ki)
```

```python
#@title ## 期を算出（文字列型の引数）

# Ex：kemco_ki('2025-4-1')
def ki_str(ymd):
    buf = ymd.split('-')
    year = int(buf[0])
    month = int(buf[1])
    if year >= 2025 and month >=4:
        ki= year - 1978
    elif month < 10 :
        ki = year - 1979
    else :
        ki = year - 1978

    return ki
```

```python
#@title ## 月末日を取得
import datetime
import calendar

# Ex：kemco_lastdate(2025, 6)
def kemco_lastdate(year, month):
    return datetime.date(year, month, calendar.monthrange(year, month)[1])
```

```python
#@title ## 指定した年月を元に、過去12カ月分の年月を古い順で取得する

'''
   Parameters:
       year (int)   : 年（例：2025）
       month (int)  : 月（例：6）
       n(int)       : 回数（例：12）

    Returns:
        List[str]: "YYYY-MM" 形式の年月のリスト
'''
from datetime import datetime
from dateutil.relativedelta import relativedelta
def kemco_gobackmonth(year: int, month: int, n: int):
    base_date = datetime(year, month, 1)
    return [
        (base_date - relativedelta(months=i)).strftime("%Y-%m")
        for i in reversed(range(n))
    ]

# 使用例
months = kemco_gobackmonth(2025, 6, 12)
print(months)

```

```python
#@title ## フォルダ内のCSVファイルを結合
import csv
from pathlib import Path

drive_path = '/content/drive/MyDrive/'
read_path = drive_path + '★開発中/Tips/売上データ/'
write_path = 'margin_test.csv'

rows = []
skip_num = 0    # 最初のファイルはヘッダーも読み込む
for file in Path(read_path).glob('*.csv'):
    print(file)

    f = open(file, encoding='cp932')
    reader = csv.reader(f)
    for row in reader:
        if reader.line_num <= skip_num:
            continue
        rows.append(row)
    f.close()
    skip_num = 1    # ２周目以降のループで読み飛ばすヘッダー行数

# 書き込み
f = open(write_path, mode='w', newline='', encoding='cp932')
writer = csv.writer(f)

for row in rows:
    writer.writerow(row)
f.close()

print('\n', '「margin_test.csv」に結合しました！')
```

```python
#@title ## ダウンロード（分割で書込み）
import requests
import os

url = 'https://www.pref.hiroshima.lg.jp/uploaded/attachment/627221.xlsx'
save_name = os.path.basename(url)

r = requests.get(url, stream=True)
with open(save_name, 'wb') as f:
    for chunk in r.iter_content(chunk_size=1024):
        if chunk:
            f.write(chunk)
            f.flush()
```

```python
#@title ## yahooニュースを取得（RSS） ＆ 保存
import requests
from bs4 import BeautifulSoup
from pandas import DataFrame

url = 'https://news.yahoo.co.jp/rss/topics/top-picks.xml'
ret = requests.get(url)
html = BeautifulSoup(ret.text, 'xml')
for item in html.find_all('item'):
    print(item.title.string)
    print(item.link.string)
    print()

# タグ付き保存
with open('yahoo.html', mode="w") as f:
    f.write(str(html))

# タグなし保存
with open('yahoo.txt', mode="w") as f:
    f.write(html.text)

print('OK!')
```

```python
#@title ## 最下行の取得

import openpyxl as px

path = '/content/drive/MyDrive/★開発中/Tips/data/要介護認定者数.xlsx'

#
#   wb          ：ワークブック
#   sheet_name  ：シート名
#   col         ：何列目をチェックするか（1～N）
#
def excel_max_row(wb, sheet_name, col):
    ws = wb[sheet_name]
    maxrow = ws.max_row + 1
    maxcol = ws.max_column + 1

    for j in range(1,maxcol):
        if j == col:
            for i in reversed(range(1,maxrow)):
                if ws.cell(row=i, column=j).value != None:
                   # print(str(j) + '列目の最終行：' + str(i))
                    break
    return i

wb = openpyxl.load_workbook(path)
col = 1
max = excel_max_row(wb, '第２表h', col)
print(str(col) + '列目の最下行：' + str(max))
```

```python
#@title ## 今日の日時
import datetime
from pytz import timezone

# Google Colabでは協定世界時をデフォルトにしているため、これを日本時間に直す必要があります。
dt = datetime.datetime.now()   # 現在時刻生成
dt = dt.astimezone(timezone('Asia/Tokyo'))    # タイムゾーンを東京に

year = dt.year
month = dt.month
day = dt.day
hour = dt.hour
minute = dt.minute
second = dt.second

# 曜日
days_japanese = ['月曜日', '火曜日', '水曜日', '木曜日', '金曜日', '土曜日', '日曜日']
weekday_name =dt.strftime('%A')  # これだと英語表記（参考）
weekday_jp = days_japanese[dt.weekday()]

print(f'{year}年{month:02}月{day:02}日({weekday_jp})')
print(f'{hour:02}時{minute:02}分{second:02}秒')
```

```python
#@title ## 数値をゼロ埋め（ゼロパディング）

# 数値「12」を文字列「0012」に書式化
num = 12

# f文字列：Python 3.6以降
s = f'{num:04}'  # 0埋めで4文字
print(s)  # 0012

# formatメソッド：Python 2.6以降（3を含む）
# formatメソッドに渡した位置引数が{0}、{1}、……に渡される
s = '{0:04}'.format(num)  # 第0引数：0埋めで4文字
print(s)  # 0012

# format関数：Python 2.6以降（3を含む）
s = format(num, '04')  # 0埋めで4文字
print(s)  # 0012

# %演算子
s = '%04d' % num  # 0埋めで4文字、10進整数
print(s)  # 0012
```
